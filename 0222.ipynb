```python
# 비공개 속성 사용하기
# 인스턴스.속성 = 값을 통해
# 새롭게 인스턴스의 속성 값을 변경 못하도록 막고 싶을 때는 비공개 속성을 사용하면 됩니다.
# 사용법은 해당하는 속성 앞에 던더(__) 언더바언더바를 붙이면 비공개 속성이 됩니다.

class ClassName:
    def __init__(self, param1, param2):
        self.attr1 = param1
        self.__attr2 = param2 # 비공개 속성
```

##### 비공개 속성



```python
class Person:
    def __init__(self, name, age, address, wallet):
        self.name = name
        self.age = age
        self.address = address
        self.__wallet = wallet # 변수 옆에 __를 붙여서 비공개 속성으로 만듦
    def pay(self, amout):
        if self.__wallet < amout:
            print("돈이 모자랍니다.")
            return
        self.__wallet -= amout # 비공개 속성은 클래스 안의 메서드에서만 접근할 수 있음
        print("이제 {0}원 남았네요".format(self.__wallet))
    
maria = Person("마리아", 20, "서울시 서초구 반포동", 10000)
maria.pay(3000)
```

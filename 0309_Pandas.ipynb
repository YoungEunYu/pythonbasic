```python
pokemons = pd.read_csv('ignore/pokemon.csv', index_col=['Pokemon']).squeeze()
pokemons
```

```python
def single_or_multi(type):
    if '/' in type:
        return 'Multi'
    return 'Single'


# 시리즈 apply() 메서드
# 시리즈의 각 값에 대해 모두 함수를 적용하는 방법.
# 시리즈 객체의 값에 대해 한 번씩 함수를 호출하고 함수 호출의 반환 값으로 구성된 새로운 시리즈 객체를 반환한다.

pokemons.apply(single_or_multi)
```

```python
pokemons.apply(single_or_multi).value_counts()
```

```python
import seaborn as sns
titanic = sns.load_dataset("titanic")
titanic['age'].fillna(titanic['age'].mean(), inplace=True)
bins=[1, 20, 30, 50, 70, 100]
labels = ['미성년자', '청년', '장년', '중년', '노년']

titanic['연령층'] = pd.cut(titanic['age'], bins=bins, labels=labels)
titanic['category3'] = titanic.apply(lambda i: '미성년자' if i.age < 20 else i.연령층
                                    +('남성' if i.sex == 'male' else '여성'), axis=1)
titanic[['연령층', 'age', 'sex', 'category3']]
```

```python
# 해당되지 않는 애들 조회, 1보다 작은 애들
# 버티컬 바 | == true, false = true / false, false = false

titanic[(titanic['age']<1) | (titanic['age'] > 100)][['연령층', 'age', 'sex', 'category3']]
```

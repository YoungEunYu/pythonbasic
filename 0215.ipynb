여러가지 조건을 일일히 체크해야 할때는 if 여러개 쓰는게 맞다.
그런데 if가 아니면 다른 조건 비교, 아니면 또 다른 조건 비교
이렇게 비교하려면 if elif elif 로 사용한다. 

```python
a = int(input("발주 수량을 입력하세요"))
print(a, "개를 주문하셨습니다.", sep="")

if a < 1000:
    dc = 0
elif 1500 > a > 1000:
    dc = 5/100
elif 2000 > a > 1500:
    dc = 1/10
elif a > 2000:
    dc = 2/10


print("적용 할인율은 ", int(dc*100), "%입니다.", sep="")

print("총 발주 금액은 $", int(a*10 - (a*dc)), "입니다.", sep="")
```

```python
# pandas는 두개 이상의 데이터프레임을 하나로 합치는 병합 merge와 연결 concatenate를 지원한다
```

```python
# merge 함수는 두 데이터 프레임의 공통 column 혹은 인덱스를 기준으로 두 개의 테이블을 합칩니다.
# 이 때 기준이 되는 column, row의 데이터를 키(key)라고 합니다.
```

```python
df1 = pd.DataFrame({'고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
                   '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']},
                   columns = ['고객번호', '이름'])
df1
```

```python
df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]},
    columns = ['고객번호', '금액'])
df2
```

```python
# 합쳐진다
pd.merge(df1, df2)
```

```python
# 겹치는 고객번호만 돌려준다
pd.merge(df1, df2, how = 'inner')
```

```python
# outer join 방식은 키 값이 한쪽에만 있어도 데이터를 보여줍니다.
# 고객번호 안겹쳐도 돌려준다

pd.merge(df1, df2, how = 'outer')
```

```python
# left는 첫 번째 인수 기준으로, right는 두번째 인수 기준으로 DataFrame의 키 값을 모두 보여줍니다.

pd.merge(df1, df2, how='left')
```

```python
pd.merge(df1, df2, how='right')
```

```python
# 만약 테이블에 키 값이 같은 데이터가 여러개 있는 경우에는
# 있을 수 있는 모든 경우의 수를 따져서 조합을 만들어 냅니다.

df1 = pd.DataFrame({
    '품종': ['se', 'se', 'vir', 'vir'],
    '꽃잎길이': [1.4, 1.3, 1.5, 1.3]},
    columns = ['품종', '꽃잎길이'])

df1
```

```python
df2 = pd.DataFrame({
    '품종': ['se', 'vir', 'vir', 'versi'],
    '꽃잎너비': [0.4, 0.3, 0.5, 0.3]},
    columns = ['품종', '꽃잎너비'])

df2
```

```python
# 이 데이터에서 키 값 setosa에 대해 왼쪽 DataFrame는 1.4와 1.3라는 2개의 데이터,
# 오른쪽 DataFrame에 0.4라는 1개의 데이터가 있으므로
# 병합된 데이터에는 setosa가 (1.4, 0.4), (1.3, 0.4) 두 개의 데이터가 생깁니다.
# 키 값 virginica의 경우에는 왼쪽 DataFrame에 1.5와 1.3라는 2개의 데이터,
# 오른쪽 DataFrame에 0.3와 0.5라는 2개의 데이터가 있으므로 2개와 2개의 조합에 의해 4가지 값이 생깁니다. 


pd.merge(df1, df2)
```

```python
# 그룹분석

np.random.seed(0)
df2 = pd.DataFrame({
    'key1': ['A', 'A', 'B', 'B', 'A'],
    'key2': ['one', 'two', 'one', 'two', 'one'],
    'data1': [1,2,3,4,5],
    'data2': [10, 20, 30, 40, 50]})
df2
```

```python
# 조회하면 데이터프레임 그룹바이로 되어 있다.

groups = df2.groupby(df2.key1)
groups
```

```python
groups.groups
```

```python
# a그룹과 b그룹 데이터의 합계를 구하기 위해 sum이라는 그룹연산을 합니다.
groups.sum()
```

```python
# GroupBy 클래스 객체를 명시적으로 얻을 필요가 없다면 groupby 메서드와 그룹연산 메서드를 연속으로 호출합니다.
# 다음 예제는 column data1에 대해서만 그룹연산을 하는 코드입니다.

groups = df2.data1.groupby(df2.key1).sum()
groups
```

```python
# 똑같은 결과!
df2.groupby(df2.key1).sum()['data1']
```

```python
# 복합 키 (key1, key2) 값에 따른 data1의 합계를 구하자. 분석하고자 하는 키가 복수이면 리스트를 사용합니다.

df2.data1.groupby([df2.key1, df2.key2]).sum()
```

```python
# BMS: 주말이 아닌 평일 중에서 각 달의 첫날
pd.date_range('2018-1-1', '2018-12-31', freq = 'BMS')
```

```python
# BM: 주말이 아닌 평일 중에서 각 달의 마지막 날 
pd.date_range('2018-1-1', '2018-12-31', freq = 'BM')
```

```python
# WOM-2THU: 각 달의 두번째 목
pd.date_range('2018-1-1', '2018-12-31', freq = 'BM')
```

```python
# 시계열 데이터의 인덱스는 시간이나 날짜를 나타내기 때문에 날짜 이동 등의 다양한 연산이 가능합니다.
# 예를 들어 shift 연산을 사용하면 인덱스는 그대로 두고 데이터만 이동할 수도 있다.

np.random.seed(0)
ts = pd.Series(np.random.randn(4), index = pd.date_range('2018-1-1', periods=4, freq='M'))
ts
```

```python
ts.shift(-1)
```

```python
ts.shift(1, freq='M')
```

```python
# ts로부터 일주일씩 미뤄진것.

ts.shift(1, freq='W')
```

```python
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.alive.unique()
```

```python
def change_boolean(value):
    if value == 'yes':
        return True
    else:
        return False

titanic.alive = titanic.alive.apply(change_boolean)
titanic
```

```python
def single_or_multi(type):
    if '/' in type:
        return 'Multi'
    return 'Single'


# 시리즈 apply() 메서드
# 시리즈의 각 값에 대해 모두 함수를 적용하는 방법.
# 시리즈 객체의 값에 대해 한 번씩 함수를 호출하고 함수 호출의 반환 값으로 구성된 새로운 시리즈 객체를 반환한다.

pokemons.apply(single_or_multi)
```
